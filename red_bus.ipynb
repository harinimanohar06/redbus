```python
data.isna().sum()
```

```python
data['bustype'] = data['bustype'].fillna(data['bustype'].mode()[0])
```

```python
data['star_rating']=data['star_rating'].fillna(data['star_rating'].median())
data['seats_available']=data['seats_available'].fillna(data['seats_available'].median())
```

```python
data.isna().sum()
```

```python
data
```

```python
import mysql.connector

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="root",
)

cursor = conn.cursor()
cursor.execute("CREATE DATABASE redbus_data")
print("Database created successfully!")
```

```python
cursor.execute("""CREATE TABLE redbus_data.busdetail (
    state VARCHAR(100),
    route_name VARCHAR(255),
    route_link TEXT,
    busname VARCHAR(255),
    bustype VARCHAR(255),
    departing_time TEXT,
    duration VARCHAR(50),
    reaching_time text,
    star_rating DECIMAL(2,1),
    price DECIMAL(10,2),
    seats_available INT
);""")
conn.commit()
```

```python
for i,row in data.iterrows():
    insert_query = """insert into redbus_data.busdetail(state,route_name,route_link,busname,bustype,departing_time,duration,reaching_time,
                       star_rating,price,seats_available)values(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"""
    values = (row['state'],row['route_name'],row['route_link'],row['busname'],row['bustype'],row['departing_time'],row['duration'],
              row['reaching_time'],row['star_rating'],row['price'],row['seats_available'])
    cursor.execute(insert_query,values)
    conn.commit()
```

```python
cursor.execute("SELECT * FROM redbus_data.busdetail")
data = cursor.fetchall()
columns = [desc[0] for desc in cursor.description]
table = pd.DataFrame(data, columns=columns)
```
